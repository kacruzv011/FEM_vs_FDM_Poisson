```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
# Cargar el CSV generado por diferencias finitas
csv_path = "data/fem_poisson_case1.csv"
df = pd.read_csv(csv_path)


# Asegurar que los valores estn ordenados correctamente
df = df.sort_values(by=["y", "x"])

# Obtener vectores nicos de coordenadas
x_vals = np.sort(df["x"].unique())
y_vals = np.sort(df["y"].unique())

# Crear grillas X, Y
X, Y = np.meshgrid(x_vals, y_vals)

# Reshapear Z para que tenga la misma forma que X e Y
Z = df["V"].values.reshape(len(y_vals), len(x_vals))

# Graficar
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k', linewidth=0.3)
ax.set_title("Solucin FEM - Caso 1")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("V(x,y)")
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.tight_layout()
plt.show()
```
